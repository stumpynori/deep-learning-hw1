# 1.-2. task

In this task I chose 5 picture of handball players which were 128x128 sized. I found a great code here: https://www.codementor.io/@innat_2k14/image-data-analysis-using-numpy-opencv-part-1-kfadbafx6
and I used this to show the RGB channels, but I do the understanding part and use my own comments. I only print out the numerical channels in the first picture, but of course it is the same for all 5.

```python
import numpy as np
import imageio
import matplotlib.pyplot as plt
from sklearn import preprocessing
pic = imageio.imread('1.jpg') #read file
fig, ax = plt.subplots(nrows = 1, ncols=3, figsize=(15,5)) #this is for the one-line plot
averages = []
stds = []
averages_n = []
stds_n = []
for c, ax in zip(range(3), ax): #c is for color, ax is for the subplot
    
    split_img = np.zeros(pic.shape, dtype="uint8") #this creates a 128 lenght zero vector, we will put the numbers in it

    split_img[ :, :, c] = pic[ :, :, c] #devide the picture for channels

    ax.imshow(split_img) #plot all the channels separately and print the arrays
    array = []
    for i in split_img: #i know it is not beutiful, but it works:D 
        for j in i:
            for z in j:
                if z != 0:
                    array.append(z) #i got the real values of each channel
    print(array) #print out the RGB channels, I only used it here, but it is the same for all pictures
    average = np.average(array) #it counts the average and standard deviation of the lists, and put it in the output
    std = np.std(array)
    averages.append(average)
    stds.append(std)
    array_normed = [(i-average)/std for i in array] #normalize the list, and do the average and std part with it
    average_n = np.average(array_normed)
    std_n = np.std(array_normed)
    averages_n.append(average_n)
    stds_n.append(std_n)
    
print(averages, stds, "\n", np.round(averages_n,1), np.round(stds_n,1)) #print out the averages and stds for the real and the normalized arrays
    


    
```

# 3. task 

I've never worked with audio in Python before, so I used some code what I found here: https://pythontic.com/visualization/signals/spectrogram
I read that the spectrogram can be visualized if the file is a mono wav file, so I download one from this website: https://www.ee.columbia.edu/~dpwe/sounds/music/
After this I download a music in mp3 and convert into wav and I investigate the differences between the two music.
For third step I was courious what is the difference in spectrograms when we take an exact sound, this case a bike honk.
I really like this task :)

```python
import matplotlib.pyplot as plot

from scipy.io import wavfile

 

# Read the wav file (mono)

samplingFrequency, signalData = wavfile.read('africa-toto.wav')

# Plot the signal read from wav file

plot.subplot(211)
plot.title('Spectrogram of a wav file')
plot.plot(signalData)
plot.xlabel('Sample')
plot.ylabel('Amplitude')

 

plot.subplot(212)
plot.specgram(signalData,Fs=samplingFrequency)
plot.xlabel('Time')
plot.ylabel('Frequency')

plot.show()
```

```python
from urllib.request import urlopen
import bs4
url = 'https://en.wikipedia.org/wiki/Deep_learning'
source = urlopen(url).read().decode('utf-8')
soup = bs4.BeautifulSoup(source, "lxml")
```

```python
url = 'https://en.wikipedia.org/wiki/Deep_learning'
html = urlopen(url).read()
soup = bs4.BeautifulSoup(html)

for script in soup(["script", "style"]):
    script.decompose()


strips = list(soup.stripped_strings)
print(strips[:10])
```

```python
from collections import Counter
import re
txt =""
numero=[]
for i in strips:
    txt += i
print(txt)
```

```python
alphabet=['a','b','c','d','e','f','g','h','i','j','k','l','m','n','o','p','q','r','s','t','u','v','w','x','y','z']
for i in alphabet:
    count = len(re.findall(i, txt))   
    numero.append(count)
```

```python
plt.bar(alphabet,numero)
plt.title('Number of letters on the website')
plt.xlabel('letters')
plt.ylabel('frequency')
plt.show()
```
